```python
import requests
import re 
from bs4 import BeautifulSoup 
```

```python
URL = "https://www.hundeweb.dk/dkk/public/openIndex?ARTICLE_ID=153&RAID=2620&AAR=2024"
r = requests.get(URL)
soup = BeautifulSoup(r.content,'html5lib')
```

```python
spk_raw_urls = []
shows = soup.find_all(string=re.compile('.*Spidshundeklubben.*'))
for show in shows:
    spk_raw_urls.append(show.parent.parent.parent.find(string='Resultater').parent['href'])
    print(show.parent.parent.parent.find(string='Resultater').parent['href'])
```

```python
raid_dict = {}
raid = soup.find(id='RAID')
for opts in raid.find_all('option'):
    raid_dict[list(opts.stripped_strings)[0].lower()] = opts['value']
    print(opts['value'],list(opts.stripped_strings)[0].lower())
```

```python
spk_raw_urls[0] + f"search?raid={raid_dict['japansk spids']}"
```
